For ratings:
**1. Movies for All (Suitable for All Ages):**
   - G (General Audiences): Suitable for all ages.
   - PG (Parental Guidance Suggested): Some material may not be suitable for children.
   - Approved

**2. Restricted Movies (Age Restrictions Apply):**
   - PG-13 (Parents Strongly Cautioned): Some material may be inappropriate for children under 13.
   - R (Restricted): Restricted to viewers over 17 or 18, or 21 in some areas, unless accompanied by an adult.

**3. Suitable but Guidance Needed (Parental Guidance Advised):**
   - TV-PG (TV Parental Guidance Suggested): Content may be suitable for children ages 10 and older. Parents are urged to provide "parental guidance."
   - TV-14 (Parents Strongly Cautioned): Content may be suitable for children ages 14 and older. Parents are urged to exercise caution and provide guidance.

**4. Adults Only (Explicit Content for Adults):**
   - NC-17 (Adults Only): No one 17 and under admitted.
   - TV-MA (Mature Audience Only): Intended for mature audiences. Content may be explicit, including strong language, sexual content, and graphic violence.

**5. Not Rated (Uncategorized or Content Varies Widely):**
   - Unrated
   - Not Rated

### Ratings

```python
df['rating'].unique()
```

```python
categories = {
    0: ['G', 'PG', 'Approved'],  #'movies_for_all'
    1 : ['PG-13', 'R'],        # 'age_restrictions'
    2 : ['TV-PG', 'TV-14'],  # 'parental_guidance_advised'
    3: ['NC-17', 'TV-MA' , 'X'],       # 'adults_only'
    4 : ['Unrated', 'Not Rated']    # 'not_rated'
}
```

```python
# Create a mapping dictionary to map ratings to numerical values
mapping = {}
for category, values in categories.items():
    for value in values:
        mapping[value] = category

# Map the 'Rating' column to numerical values based on the custom categories
df['Rating Category'] = df['rating'].map(mapping)
```

```python
df['Rating Category'].unique()
```

```python
df['Rating Category'] = df['Rating Category'].astype(dtype = np.int_)
```

```python
movie_categories = { 0 : ['Drama' , 'Biography', 'History'] , #'dramatic'
                    1 : ['Adventure','Action','Crime','Thriller','Mystery','Horror'], #'action_adventure'
                    2 : ['Comedy','Family'], #'comedy'
                    3 : ['Fantasy' , 'Sci-Fi'], #'fantasy_sci_fi'
                    4 : ['Animation','Romance','Music','Western','Sport','Musical'] #'miscellaneous'
                   }
```

```python
# Create a mapping dictionary to map ratings to numerical values
mapping_2 = {}
for category, values in movie_categories.items():
    for value in values:
        mapping_2[value] = category

# Map the 'Rating' column to numerical values based on the custom categories
df['Movie Category'] = df['genre'].map(mapping_2)
```
